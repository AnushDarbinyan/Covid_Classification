# Covid classification from lung X-rays

ներմուծվում են համապատասխան գրադարանները

```python
import os
import numpy as np
import cv2         
import matplotlib.pyplot as plt
import pickle     # модуль сериализации и десериализации объектов в Питоне для последующей их передачи.
import random
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
shape=(256,256)       #  Քանի որ նկարների չափերը կարող են տարբեր լինել, առաջադրվում է նկարի չափը, բոլոր նկարները հետագայում բերվելու է նույն չափի նոր տրվեն մոդելին
data = []
categories = ['COVID19','NORMAL']


```

Տրվում է X-Rays նկարների հղումը 

```python
dir ='D:\\Python_ML_Math_2021\\project\\lung_xray\\train'
```

```python
for category in categories:   # հերթով անցնում է categories-ի էլեմենտների վրայով և ստեղծվում է train.pickle, որը իրենից ներկայացնում է train data-յի բոլոր նկարների pickle դարձրած ֆայլը
    path = os.path.join(dir,category)
    label =  categories.index(category)
    for image in os.listdir(path):
        imgpath = os.path.join(path,image)
        print(imgpath)
        train_img = cv2.imread(imgpath,0)
        train_img=cv2.resize(train_img,shape)
        cv2.imshow('image', train_img)
        train_img = np.array(train_img).flatten()
        data.append([train_img,label])

pick_in = open('train.pickle','wb')
pickle.dump(data,pick_in)
pick_in.close()
pick_in
```

```python
features = []
labels = []
pick_in = open('train.pickle', 'rb')
data = pickle.load(pick_in)
pick_in.close()
for feature, label in data:
    features.append(feature)
    labels.append(label)



```

Նույն գործողությունները կատարում եմ Test data-յի համար

```python
dir ='D:\\Python_ML_Math_2021\\project\\lung_xray\\test' 
data2 = []
```

```python
for category in categories:
    path = os.path.join(dir,category)
    label =  categories.index(category)
    for image in os.listdir(path):
        imgpath = os.path.join(path,image)
        print(imgpath)
        test_img = cv2.imread(imgpath,0)
        test_img=cv2.resize(test_img,shape)
        cv2.imshow('image', test_img)
        test_img = np.array(test_img).flatten()
        data2.append([test_img,label])
```

```python
pick_in = open('test.pickle','wb')
pickle.dump(data2,pick_in)
pick_in.close()
pick_in


    
```

```python
pick_in = open('test.pickle', 'rb')
data2 = pickle.load(pick_in)
pick_in.close()
features_test = []
labels_test = []
for feature, label in data2:
    features_test.append(feature)
    labels_test.append(label)
```

```python
data2
```

```python
xtrain, xtest, ytrain, ytest = features, features_test, labels, labels_test
C=3
d=3
```

```python
features_test
```

```python
model = SVC(C=C, kernel='linear', gamma='auto', degree=d, verbose=True).fit(xtrain, ytrain)
```

```python
filename = 'finalized_model.sav'
pickle.dump(model, open(filename, 'wb'))

accuracy = model.score(xtest, ytest)
print("accuracy", C, " ", d, ' ', accuracy * 100, '%')
```

```python
ytest
```

```python
model = pickle.load(open(filename, 'rb'))
result = model.score(xtest,ytest)
print(result)
```

```python
img=cv2.imread('../lung_xray/test/COVID19/person100_bacteria_478.jpeg',0) # փորձարկելու համար կարելի է թոքերի X-rays-ի նկար (տալով ճիշտ հղումը) քաշել ինտերնետից և ստուգել մոդելը
tstimg=cv2.resize(img,shape)
tstimg = np.array(tstimg).flatten()
tstimg=tstimg.reshape(1,-1)
prediction = model.predict(tstimg)
cv2.resize(img,(800,600))
if prediction:
    cv2.putText(img,'N E G A T I V E',(100,100),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),3)
else:
    cv2.putText(img,'P O S I T I V E',(300,300),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),3)

cv2.imshow('hey',cv2.resize(img,(500,500)))
cv2.waitKey(-1)
```
